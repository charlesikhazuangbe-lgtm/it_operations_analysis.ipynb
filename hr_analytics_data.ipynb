HR Analytics Project

Introduction

Using SQL, Python (Pandas), and Power BI to analyze employee salary, promotion status, and departmental trends. Generated insights on average salary by department, promotion distribution, and workforce metrics through interactive dashboards.

Objective to show:

1. Salary by Department
2. Total Employees 
3. Average Salary 
4. Promotion Status Breakdown

Load HR CSV File

```python
import pandas as pd
df=pd.read_csv('hr_analytics_data.csv')
df.head()
```

Total Employees

```python
unique_count = df['employee_id'].nunique()
print(unique_count)
```

Average Salary by Department

```python
df.groupby('department')['salary'].mean()
```

Highest-Paid Employee

```python
df.loc[df['salary'].idxmax()]
```

Promotion status counts

```python
df['promotion_status'].value_counts()
```

Salary by Department (Bar Chart)

```python
import matplotlib.pyplot as plt
df.groupby('department')['salary'].mean().plot(kind='bar')
plt.title('Average Salary by Department')
plt.xlabel('department')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

Promotion Status Chart

```python
import matplotlib.pyplot as plt
df['promotion_status'].value_counts().plot(kind='bar')
plt.title('Promotion Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Salary Distribution

```python
import matplotlib.pyplot as plt
df["salary"].plot(kind="hist")
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.show()
```
